### Determinining convective masses

```python
from terra import pacha
```

terra can calculate the convective mass of a given star using only its metallicity ([Fe/H]) and mass. For example, the convective mass of the metal rich star HIP 71726 can be estimated as:

```python
pacha.cvmass(feh=0.164, mass=1.18)
```

pacha makes a double interpolation between the [Fe/H] and mass. The convective masses (in solar units) adopted for this calculation are from http://www.astro.yale.edu/yapsi/. 

It is important to highlight that the valid ranges of [Fe/H] and mass to determine convective masses are: 

-1.0 dex <= [Fe/H] <= 0.3 dex and 0.5 M_sun <= mass <= 1.3 M_sun

### modeling planet engulfment

For a detailed explanation of how terra works, please see our paper at https://ui.adsabs.harvard.edu/abs/2016A&A...589A..65G. Unlike the old terra code, this updated version uses the new solar abundances of Asplund (https://ui.adsabs.harvard.edu/abs/2021A%26A...653A.141A/abstract). 

In this tutorial, we will modelate the observed pattern abundance of HIP 71726 using terra. For that, we need an input table with all the observed abundances of HIP 71726. Download the 'obs_abd.csv' table from terra github, then run the following command.

```python
pacha.mod_abd(feh=0.164, mass=1.18, data_input='obs_abd.csv')
```

terra will create the plots observed above and also two csv tables. The first is the 'chi2_convolution.csv', which contains the number of rocky and terrestrial material needed to reproduce the observed abundance pattern of HIP 71726. This value is slightly lower than that reported in our paper (https://ui.adsabs.harvard.edu/abs/2021arXiv210900679G/abstract), but still consistent. terra also creates a table containing the predicted abundances ('results.csv').

Finally, make sure the column 'element' in your inpu_table has no white spaces, e.g., 'C ', 'Si ' because terra can ignore these elements. To see which elements are available for terra, please see the table 'data/seab_data.csv' in GitGub.
